```python
# alphasV3       This function calculates the fraction of free ions in a
#                solution of fixed composition and ionic strength (given)
#                needed for correcting free activity coefficients for ion
#                complexation. 
#                    
#               [alpha_cat, alpha_an, iter] = alphasV3(mT_cat, mT_an)
#                     alpha_cat  .......... [M]f/[M]T
#                     alpha_an   .......... [X]f/[X]T
# 
#
#                Assumptions:
#                1) 1:1 complexes only. Higher order complexes ignored
#                2) Trivalents form strong hydroxyl complexes
#                3) Br/Cl- ion pairs weak therefore ignored
#                           AlphaCl = 1.0  or [Cl]f = [Cl]T
#                           AlphaBr = 1.0  or [Br]f = [Br]T
#                5) Assume K forms only KSO4- Complex
#                6) The speciation of minor components of seawater can be
#                   made without making any iterations, since minor components
#                   do not affect the speciation of the major componenents 
#                
#
#                NOTE: The model is only good as the K data inputed.          
#
#              Major Cations                  Minor Cations 
#              Na, K, Mg, Ca, Sr              H, Li, Rb, Cs, NH4, Ba, Mn, Fe(II), Co, Ni, Cu(II), Zn
#                                             UO2, Be, Cd, Pb(II), MgOH, (MgF),(CaF) Cu(I)
#                                
#                                             REE
#                                             La, Ce, Pr, Nd, Pm, Sm, Eu, Gd, Tb, Dy, Ho, Er, Tm, Yb, 
#                                             Lu, Y
#                               
#                                             Trivalent
#                                             Al, Ga, In, Fe(III)
#
#
#             Major Anions                    Minor Anion
#             Cl, SO4, CO3, HCO3              HSO4, HS, OH, I, ClO3, ClO4, BrO3, CNS, NO2, NO3,
#             Br, F, B(OH)4                   H2PO4, HPO4, PO4, H2AsO4, HAsO4, AsO4, HSO3, SO3, Acetate
#
#               
#             'For major cations, the calculations a made using an iterative process
#              For minor components of seawater, calculations can be made without making 
#              any iteration, since minor components do not affect the speciation Of the 
#              major component.'    .....  Millero, 2006       
#               

## ----------------------------------------------------------------------------
##  Assign values to global variables
mjCaMax = 5         #In python array goes 0,1,2,3,4,5,6, so mjCaMax must go to 7 to include the 6th variable
mnCaMax = 37
CaMax = mjCaMax + mnCaMax

mjAnMax = 7          #In python array goes 0,1,2,3,4,5,6,7,8 so mjCaMax must go to 8 to include the 7th variable
mnAnMax = 19
AnMax = mjAnMax + mnAnMax

NeutMax = 7

def alphasV3(Tc, I, mT_cat, mT_an, OH):
    
    ## Determine lengths of input vectors
    veclengths=[size(Tc), size(I), size(len(mT_cat)), size(len(mT_an))]
    maxr = max(veclengths)
    
    if len(unique(veclengths)) > 2:
        print('*** INPUT ERROR: Input vectors must all be of same length, or of length 1. ***')
        
## ----------------------------------------------------------------------------
    mjKcond, mnCat_Kcond, divKcond, trivKcond  = Kcond(Tc, I, mT_cat,mT_an)
    alpha_cat2 = np.zeros([maxr,CaMax])
    alpha_an2  = np.zeros([maxr,AnMax])
    mT_an.append(OH)      # 10^(-14+pH)   

## MAJOR CAT-AN
## Iteration until convergence      

    alpha_cat2[0][0:5] = 1              # Initialize iteration by setting all alpha_cat = 1 
    alpha_an2[0][0:8] = 1               # Initialize iteration by setting all alpha_an = 1  

    itera = 2    
    error1 = 1
    error2 = 1
    tolerance = 0.00001

    alpha_an1, alpha_cat1 = guess_alphaV3(alpha_an2, alpha_cat2, mT_cat, mT_an, mjKcond, maxr)

    saa1 = sum(any(abs(error1 > tolerance)))
    saa2 = sum(any(abs(error2 > tolerance)))

    if saa1 > 0 or saa2 > 0:
        alpha_an, alpha_cat = guess_alphaV3(alpha_an1, alpha_cat1, mT_cat, mT_an, mjKcond, maxr)
        error1 = alpha_cat - alpha_cat1
        error2 = alpha_an - alpha_an1
        while itera < 9:
            itera = itera + 2
            
    mF_an = alpha_an[0][0:8]* mT_an
    mF_cat = alpha_cat[0][0:5]* mT_cat
    
    ## Minor Cations
    Lig = np.zeros([1,7])
    Lig = [1,2,3,4,5,6,8]           #[Cl, SO4, CO3, HCO3, Br, F, OH]

    A = np.zeros([maxr,7])
    for i in range(maxr,maxr+1):
        for II in range(7):
            for JJ in range(7):
                for KK in range(3):
                    if  mnCat_Kcond[KK, II, JJ] !=0:
                        A[0][II] = A[0][II] +  mnCat_Kcond[KK, II, JJ] * (mF_an[Lig[JJ]-1]**(KK+1))
                    
            
    alpha_cat[0][5:11] = 1/(A[0][0:6] + 1) 
    alpha_cat[:,21] = 1/(A[:,6] + 1)
                                                                      
    ## DIVALENTS 
    L = np.zeros([1,6])
    L = [1,2,3,4,6,8]             # [Cl, SO4, CO3, HCO3, F, OH]

    B = np.zeros([maxr,10])
    for i in range(maxr,maxr+1): 
        for II in range(10):
            for JJ in range(6):
                for KK in range(4):
                    B[0][II] = B[0][II] + divKcond[KK, II, JJ] * (mF_an[L[JJ]-1]**(KK+1))   
                

    alpha_cat[0][11:21] = 1/(B + 1) 

    ## TRIVALENT
    
    C = np.zeros([maxr,20])
    for i in range(maxr,maxr+1):
        for II in range(20):
            for JJ in range(6):
                for KK in range(4):
                    C[0][II] = C[0][II] + trivKcond[KK, II, JJ] * (mF_an[Lig[JJ]-1]**(KK+1))

    alpha_cat[0][22:42] = 1/(C + 1)
    
    return alpha_cat, alpha_an, mF_an, mF_cat, itera

```

```python
a_cat, a_an, mF_an, mF_cat, itera = alphasV3(25, 0.7228, [0.4689674, 0.0102077, 0.0528171, 0.0102821, 0.0000907], [0.5458696, 0.0282352, 0.00024694, 0.00179143, 0.0008421, 0.0000683, 0.0001008], 10**(-14+8.1))
print(a_cat)
```
