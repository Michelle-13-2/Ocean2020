```python
##  Assign values to global variables
mjCaMax = 5         #In python array goes 0,1,2,3,4,5,6, so mjCaMax must go to 7 to include the 6th variable
mnCaMax = 37
CaMax = mjCaMax + mnCaMax

mjAnMax = 7         #In python array goes 0,1,2,3,4,5,6,7,8 so mjCaMax must go to 8 to include the 7th variable
mnAnMax = 19
AnMax = mjAnMax + mnAnMax

NeutMax = 7

## --------------------------------------------------------------------------------------
def speciation (Tc, I, mT_cat, mT_an, OH):
    
    alpha_cat, alpha_an, mF_an, mF_cat, itera  = alphasV3(Tc, I, mT_cat,mT_an, OH)    
    mjKcond, mnCat_Kcond, divKcond, trivKcond  = Kcond(Tc, I, mT_cat,mT_an)

    ## Determine lengths of input vectors   
    veclengths=[size(Tc), size(I), size(len(mT_cat)), size(len(mT_an))]
    maxr = max(veclengths)
    
    if len(unique(veclengths)) > 2:
        print('*** INPUT ERROR: Input vectors must all be of same length, or of length 1. ***')
    
    ## Major cation speciation 
    mjcat_spec = np.zeros([maxr*mjCaMax,8])

    for i in range(maxr,maxr+1):
        for cat in range(5):
            for an in range(8):
                mjcat_spec [cat,an] = mjKcond[cat, an]* mF_an[an]* alpha_cat[:,cat]
                          
    sz = [1,maxr*mjCaMax]
    A = reshape(alpha_cat[0][0:5],sz)
    mjcat_FREE = A * 100
    # mjcat_percent = mjcat_spec * 100
      
    ## major anion speciation  
    mjan_spec = np.zeros([8,maxr*mjCaMax])     
    
    for an in range(8):
        for cat in range(5):
            mjan_spec [an,cat] = mjKcond[cat,an]* mF_cat[cat]* alpha_an[:,an]
                
    sz = [1,8]
    B = reshape(alpha_an[0][0:8],sz)
    mjan_FREE = B * 100
    # mjan_percent =   mjan_spec * 100
                
    ## Minor Cations
    Lig = np.zeros([1,7])
    Lig = [1,2,3,4,5,6,8]           #[Cl, SO4, CO3, HCO3, Br, F, OH]
    
    C = np.zeros([1,7])
    C = [6,7,8,9,10,11,22]          #[Li, Rb, Cs, NH4, Ba, Mn, La]

    mncat_spec = np.zeros([3,7,7])
                
    for II in range(7):
        for JJ in range(7):
            for KK in range(3):
                if mnCat_Kcond[KK, II, JJ] !=0: 
                    mncat_spec[KK, II, JJ] =  mnCat_Kcond[KK, II, JJ] * (mF_an[Lig[JJ]-1]**(KK+1)) * alpha_cat[0][C[II]-1]

    D = np.zeros([1,maxr*7])
    D[0][0:6] = alpha_cat[0][5:11]
    D[:,6] = alpha_cat[:,21]        
    mncat_FREE = D * 100
    # mncat_percent =  mncat_spec * 100
    
    ## DIVALENTS 
    L = np.zeros([1,6])
    L = [1,2,3,4,6,8]               # [Cl, SO4, CO3, HCO3, F, OH]                       
    
    div_spec = np.zeros([4,10,6])    
    
    for II in range(10):
        for JJ in range(6):
            for KK in range(4):
                if divKcond[KK, II, JJ] != 0:
                    div_spec [KK, II, JJ] = divKcond[KK, II, JJ] * (mF_an[L[JJ]-1]**(KK+1)) * alpha_cat[0][11+II]
    
    sz = [1,10*maxr]
    E = reshape(alpha_cat[0][11:21], sz)
    div_FREE = E * 100
    # div_percent = div_spec * 100
    
    ## TRIVALENT
    triv_spec = np.zeros([4, 20, 6])
                
    for II in range(20):
        for JJ in range(6):
            for KK in range(4):
                if trivKcond[KK, II, JJ] != 0:
                    triv_spec [KK, II, JJ] = trivKcond[KK, II, JJ] * (mF_an[L[JJ]-1]**(KK+1)) * alpha_cat[0][22+II]

    sz = [1,20]
    F = reshape(alpha_cat[0][22:42],sz)
    triv_FREE = F * 100
    # triv_percent =  triv_spec *100
                
    return mjcat_spec, mjcat_FREE, mjan_spec, mjan_FREE, mncat_spec, mncat_FREE, div_spec, div_FREE, triv_spec, triv_FREE
                                                                                          
```
