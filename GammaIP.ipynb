```python
# gammaIP    This function calculates and stores activity coefficients for 
#            all ion pairs that are needed to convert thermodynamic equilibrium 
#            constants to stoichiometric constants: 
#            
#                            *******************************
#            [gmjIP, mnCatIP, gdivIP, gtrivIP] = gammaIP(Tc,I, mT_cat, mT_an) 
#                              Tc = Temperature in degrees Celcius
#                               I = Ionic strength
#                          mT_cat = Total concentration (molals) of the major 
#                                   cations in seawater (Na, K, Mg, Ca, Sr)
#                           mT_an = Total concentration (molals) of the major 
#                                   anions in seawater  (Cl, SO4, CO3, HCO3, Br, F, B(OH)4)
#                            ******************************* 
#                                                       
#            The availabity of reliable activity coefficients that can be assigned to 
#            ion pairs is a major limitation when using ion association model.
#            Unfortunately, simplifications and assumptions have to be made to fill in 
#            data gaps. Also,there is little to no data on temperature, pressure and ionic 
#            strength dependence of these activity coefficients. Therefore, we have use 
#            values estimated at 25 degrees celcius, 1 atmospheric pressure and ionic 
#            strength  of 0.7m. [Millero & Schreiber (1982)]. This, no doubt, introduces 
#            errors into the model when the values extrapolated for different ionic 
#            strengths and temperatures.
#      
#            Garrels and Thompson (1962) used the following assumptions: 
#             - 1.13 for neutral ion pairs
#             - 0.68 for +1/-1 ion pairs 
#             - 0.22 for +2/-2 ion pairs. 
#            Millero & Schreiber (1982) estimated activity coefficient using empirical equilibrium constants 
#            for a number of ion pairs at 25 degrees celcius and 0.7m . Their results were consistent 
#           (except for CaCO3 and MgCO3) within errors with activity coefficients assigned by 
#            Garrels and Thompson (1962). 
#            SHIAMI follows similar assumptions for ion pair gamma
#
#           The activity coefficients used to calculate conditional equilibrium constants 
#           in SHIAMI model for ion pairs are listed below (presently major cations and 
#           anions only) with brief discussion of reasoning behind assumptions.  
#           We try to model ionic dependence where possible  e.g via cations/anions that have similar values at 0.7m
#
## 1) For neutral ion pairs made from monovalent cations and anions...
#  An activity coefficient of 1.01 is assumed:  
#  Millero & Schreiber (1982) assign a value of 1.1 but found gNaF to be inconsistent 
#  with their assumptions. They attribute it to uncertainties in equilibrium constants used to calculate it 
#  and difficulties in estimating gamma for weakly complexed ions/ K < 1). However, their gHF, gNaHCO3, 
#  gNaB  all comform to this assumption. However, SHIAMI (and MyAMI) predicts a value of "1.01" at 
#  the same conditions which is slightly lower than that of Millero & Schreiber (1982). SHIAMI  calculates gHF as a function
#  of [Na]T (see gammaNV3.m) which in turn depends on the salinity.  As ionic strength and salinity 
#  are linked, I propose we can use it to make a crude approximation for the ionic dependence of 
#  neutral ion pairs made from monovalent cations and anions.
#
## 2) For -1 ion pairs made from monovalent cation and divalent anion...
#  An activity coefficient of 0.68(1 d.p) is assumed:.  
#  Garrels and Thompson (1962) assumed a value of 0.68 for +1/-1 ions. In good agreement, 
#  Millero & Schrieber (1982) obtain a value of 0.64. This is equal to the value they calculate 
#  for activity coefficient for Br- which they propose can be used to model ionic dependence. As SHIAMI
#  calculates a gBr- of 0.68 (1 d.p), we choose to opt for Garrels and Thompson (1962) convention. 
#
## 3) For neutral ion pairs made from divalent cations and anions...   
#  An activity coefficient of 1.0 is assumed:
#  Reardon & Langmuir (1976) suggest that " the activity coefficients of MgCO3 and CaSO4 ion
#  pairs decrease with  ionic strength (I) at 25°C" via "the empirical equation log yi = -BI. 
#  B coefficients of 0.63 +/- 0.10 for MgCO3 and 0.45 +/- 0.15 for CaSO4, are obtained from 
#  linear regression of log yL values vs I between 0.04 and 0.6 molal". They conclude that 
#  "assumptions that the activity coefficients of these neutral ion pairs equal unity or are 
#  approximated by the Setchenow expression (log yi = kI) are therefore invalid at moderate ionic 
#  strengths." Millero & Schreiber obtain a higher value (1.0) for CaSO4 than Reardon & Langmuir (1976) 
#  but find that they also obtain lower values for neutral metal-CO3 ion pairs (0.6) than prevously 
#  calculated in earlier work. But, He & Morse (1993) stated that as Pitzer parameters for carbonic 
#  acid system have been determined by assuming an activity coefficient of 1, neutral metal-CO3's activity 
#  coefficient must be assigned a value of unity. This is to limit internal inconsistents and avoid 
#  disvalidating activity coefficients calculated by the Pitzer module. Hain et al (2016) 
#  also assume unity activity coefficients for carbonate ion pairs in MyAMI model.
#
## 4) For +1 ion pairs made from divalent cation and monovalent anion...
#  An activity coefficient of  0.8  is assumed:.(except gMgF (lower), gCaF(lower), gMgOH (about right) where pitzer data is available):  
#  Garrels and Thompson (1962) assumed a value of 0.68 for +1/-1 ions whereas Millero & Schrieber (1982)
#  obtain a range of  values (0.5 - 0.9). The average, 0.7 +/- 0.1 , is in good agreement with  Garrels and Thompson (1962)
#  Like, Millero & Schrieber (1982) we choose to use the upper limit, 0.8, and the estimated  ionic dependence 
#  coefficients  [via an average of  Li+(0.9) and Na+ (0.7)] fitted to equation 27 in the paper. 
## -------------------------------------------------------------------------------------------------------

##  Assign values to global variables
mjCaMax = 5          #In python array goes 0,1,2,3,4,5,6, so mjCaMax must go to 7 to include the 6th variable
mnCaMax = 37
CaMax = mjCaMax + mnCaMax

mjAnMax = 7          #In python array goes 0,1,2,3,4,5,6,7,8 so mjCaMax must go to 8 to include the 7th variable
mnAnMax = 19
AnMax = mjAnMax + mnAnMax

NeutMax = 7

## --------------------------------------------------------------------------------------
def gammaIP(Tc,I, mT_cat, mT_an):
    
    T = Tc + 273.15 

    gF_cat, gF_an, gIP1, gdivIP, gtrivIP = gammaFV3(Tc, I, mT_cat,mT_an)
    gN = gammaNV3(Tc,I,mT_cat,mT_an)
    
    ## Determine lengths
    veclengths=[size(Tc), size(I), size(len(mT_cat)), size(len(mT_an))]
    maxr = max(veclengths)
    
    if len(unique(veclengths)) > 2:
        print('*** INPUT ERROR: Input vectors must all be of same length, or of length 1. ***')
    
    ## ------------------------------------------------------------------------------------------------------
    ## GAMMA FOR MAJOR ION PAIRS
    gmjIP = np.ones([mjCaMax, mjAnMax+1])
    
    ## 1) For neutral ion pairs made from monovalent cations and anions:
    ## 2) For -1 ion pairs made from monovalent cation and divalent anion...
    ## 3) For +1 ion pairs made from divalent cation and monovalent anion...
    
    A_phi = 3.6901532E-1 - 6.32100430E-4*T + 9.14252359/T - 1.35143986E-2*log(T) + 2.26089488E-3/(T-263) + 1.92118597E-6*(T**2) + 4.52586464e1/(680-T)   # note correction of last parameter, E+1 instead of E-1 as stated in Millero & Pierrot (1989)
    f_gamma  = -A_phi* (math.sqrt(I)/(1+1.2*math.sqrt(I)) +(2/1.2)*log(1+1.2*math.sqrt(I)))
    f_1 = 1 -(exp(-2*math.sqrt(I))*(1 + 2*math.sqrt(I) - 2*I))
    gIP_1 = exp(f_gamma + (0.3551*I) + (f_1*0.1808) + 0.0086*(I**2))  # [Millero & Schreiber 1982]
    
    ## 1) *****************************************************************************************

    gmjIP[0,0]  = gN[:,5]       # NaCl 
    gmjIP[0,3]  = gN[:,5]       # NaHCO3 
    gmjIP[0,4]  = gN[:,5]       # NaBr   
    gmjIP[0,5]  = gN[:,5]       # NaF                          
    gmjIP[0,6]  = gN[:,5]       # NaB(OH)4                    
    gmjIP[0,7]  = gN[:,5]       # NaOH 
    gmjIP[1,0]  = gN[:,5]       # KCl            
    gmjIP[1,4]  = gN[:,5]       # KBr 
    gmjIP[1,7]  = gN[:,5]       # KOH 
    
    ## 2) *****************************************************************************************
    
    gmjIP[0,1]  = gF_an[:,4]    # NaSO4                      
    gmjIP[0,2]  = gF_an[:,4]    # NaCO3                     
    gmjIP[1,1]  = gF_an[:,4]    # KSO4 
    
    ## 3) *****************************************************************************************
    
    gmjIP[2,0]  = gIP_1         # MgCl
    gmjIP[2,3]  = gIP_1         # MgHCO3        
    gmjIP[2,5]  = gIP1[:,1]     # MgF              
    gmjIP[2,6]  = gIP_1         # MgB(OH)4              
    gmjIP[2,7]  = gIP1[:,0]     # MgOH 
    gmjIP[3,0]  = gIP_1         # CaCl 
    gmjIP[3,3]  = gIP_1         # CaHCO3        
    gmjIP[3,5]  = gIP1[:,2]     # CaF                        
    gmjIP[3,6]  = gIP_1         # CaB(OH)4             
    gmjIP[3,7]  = gIP_1         # CaOH 
    gmjIP[4,0]  = gIP_1         # SrCl 
    gmjIP[4,3]  = gIP_1         # SrHCO3                    
    gmjIP[4,5]  = gIP_1         # SrF                       
    gmjIP[4,6]  = gIP_1         # SrB(OH)4                        
    gmjIP[4,7]  = gIP_1         # SrOH
    
    ## 4) For neutral ion pairs made from divalent cations and anions...  

    # gmjIP(3,2)  = 1.0;       % MgSO4            % In literature:                                  
    # gmjIP(3,3)  = 1.0;       % MgCO3            % gMgCO3    = 10 ^ (0.056 * I);      Millero & Thurmond     (1983)
    # gmjIP(4,2)  = 1.0;       % CaSO4            % gMetalCO3 =  0.6;                  Millero & Schrieber    (1982)
    # gmjIP(4,3)  = 1.0;       % CaCO3            % gCaSO4    = log yi = 10^(-0.45I);  Reardon and Langmuir   (1975)
    # gmjIP(5,2)  = 1.0;       % SrSO4            % gMgCO3    = log yi = 10^(-0.63I);  Reardon and Langmuir   (1975)
    # gmjIP(5,3)  = 1.0;       % SrCO3

    ## ************************** Minor Cat - Major An  *************************
    #  lig/An
    #  1. Cl      
    #  2. SO4    
    #  3. CO3    
    #  4. HCO3 
    #  5. Br
    #  6. F       
    #  7. OH 
    
    mjKthermo, mnCat_Kthermo, divKthermo, trivKthermo, mnAn_Kthermo = Kthermo(Tc)         # Any value for T to check if a value exists)
    mnCatIP = np.zeros([3,7,7])
    
    ## H
    mnCatIP[0,0,1] = gF_an[:,8]           # HSO4
    mnCatIP[0,0,5] = gN[:,5]              # HF
    
    ## Li
    mnCatIP[0,1,0] = gN[:,5]              # LiCl              0  +-
    mnCatIP[0,1,1] = gF_an[:,4]           # LiSO4            -1  +-2
    mnCatIP[0,1,5] = gN[:,5]              # LiF               0  +- 
    
    ## Rb 
    mnCatIP[0,2,0] = gN[:,5]              # RbCl              0  +-
    mnCatIP[0,2,1] = gF_an[:,4]           # RbSO4            -1  +-2
    mnCatIP[0,2,4] = gN[:,5]              # RbBr              0  +-
    mnCatIP[0,2,5] = gN[:,5]              # RbF               0  +-
    
    ## Cs 
    mnCatIP[0,3,0] = gN[:,5]              # CsCl              0  +-    
    mnCatIP[0,3,1] = gF_an[:,4]           # CsSO4            -1  +-2
    mnCatIP[0,3,2] = gF_an[:,4]           # CsCO3            -1  +-2
    mnCatIP[0,3,4] = gN[:,5]              # CsBr              0  +-
    
    ## NH4
    mnCatIP[0,4,1] = gF_an[:,4]           # NH4SO4           -1  +-2
    
    ## Ba
    mnCatIP[0,5,0] = gIP_1                # BaCl              +1  +2-
    mnCatIP[0,5,1] = 1.0                  # BaSO4             0  +2-2
    mnCatIP[0,5,2] = 1.0                  # BaCO3             +1  +2-
    mnCatIP[0,5,3] = gIP_1                # BaHCO3            +1  +2-
    mnCatIP[0,5,5] = gIP_1                # BaF               +1  +2-
    
    ## Cu(I)
    mnCatIP[0,6,0] = gIP1[:,3]            # Cu(I)Cl
    mnCatIP[1,6,0] = gIP1[:,4]            # Cu(I)Cl2
    mnCatIP[2,6,0] = gIP1[:,5]            # Cu(I)Cl3

    ## ------------------------------------------------------------------------------------------------------

    ## GAMMA FOR DIVALENT ION PAIRS
    #for in = 1 : maxr
    #Y3 = 10*(in-1)
    
    ## *************** ESTIMATES FOR ACTIVITY COEFFICIENTS OF SOME ION PAIRS ***************

    # matlab = [row, integer along, set]
    # python = [set, row, integer along] (keep in mind python starts at 0)

    for II in range(10):
        for JJ in range(6):
            for KK in range(4):
                if divKthermo[KK,II,JJ] !=0:                            # if ion pair exists
                    if gdivIP[KK,II,JJ] == 0:
                        if JJ in [2,3,5]:                               # model with Cu(II)
                            gdivIP [KK,II,JJ] = gdivIP[KK, 4, JJ]
                        else:
                            gdivIP [KK,II,JJ] = gdivIP[KK, 9, 0]        # PbClx 
                        
                            
    return gmjIP, mnCatIP, gdivIP, gtrivIP                           
```
