```python
#  Kcond    This function converts thermodynamic equilibrium 
#                constants (K) into conditional/stoichiometric equilibrium  
#                constants (K*)           
#                             *******************************
#            [mjKcond, mnCat_Kcond, divKcond, trivKcond] = Kcond(mjKthermo, gF_cat, gF_an, g_IP)
#      mjKcond, mnCat_Kcond, divKcond, trivKcond = thermodynamic equilbrium constants
#                                gF_cat = Free cation activity coefficients
#                                 gF_an = Free anion activity coefficients 
#                                  g_IP = Ion pair activity coefficients
#                              *******************************                         
# 
#                   In order to this, we need reliable values/estimates for: 
#
##                  1) Free ion activity coefficients(gammaFV3.m): 
#                      These are calculated as a function of temperature and ionic strength 
#                      using the Pitzer ion interaction model (gammaFV3.m)  
#
##                  2) Ion pair activity coefficients(gammaIP.m):
#                      The availabity of reliable activity coefficients that can be assigned 
#                      to ion pairs is a major limitation when using ion association model. 
#                      Unfortunately, simplifications  and assumptions have to be made to fill in 
#                      data gaps. Also,there is little to no data on temperature, pressure and 
#                      ionic strength dependence of these activity coefficients. Therefore, 
#                      we have use values estimated at 25 degrees celcius, 1 atmospheric pressure 
#                      and ionic strength of 0.7m. [Millero & Schreiber (1982)]. This, no doubt, 
#                      introduces errors into the model when the values extrapolated for different 
#                      ionic strengths and temperatures. The activity coefficients used to calculate
#                      conditional equilibrium constants in SHIAMI model are listed in appendix and 
#                      in gammaIP.m. For reasoning behind assumptions, enter "help gammaIP" 
#                      in command window or see text.
#
#                      See also gammaFV3 gammaIP
## -------------------------------------------------------------------------------------------------------

##  Assign values to global variables
mjCaMax = 5         #In python array goes 0,1,2,3,4,5,6, so mjCaMax must go to 7 to include the 6th variable
mnCaMax = 37
CaMax = mjCaMax + mnCaMax

mjAnMax = 7         #In python array goes 0,1,2,3,4,5,6,7,8 so mjCaMax must go to 8 to include the 7th variable
mnAnMax = 19
AnMax = mjAnMax + mnAnMax

NeutMax = 7

## -------------------------------------------------------------------------------------------------------

def Kcond(Tc, I, mT_cat,mT_an):

    ## Determine lengths of input vectors   
    veclengths=[size(Tc), size(I), size(len(mT_cat)), size(len(mT_an))]
    maxr = max(veclengths)
    
    if len(unique(veclengths)) > 2:
        print('*** INPUT ERROR: Input vectors must all be of same length, or of length 1. ***')
        
    ## ------------------------------------------------------------------------------------------------------
    gF_cat, gF_an, gIP1, gdivIP, gtrivIP = gammaFV3(Tc, I, mT_cat,mT_an)
    mjKthermo, mnCat_Kthermo, divKthermo, trivKthermo, mnAn_Kthermo = Kthermo(Tc)
    gmjIP, mnCatIP, gdivIP, gtrivIP = gammaIP(Tc,I, mT_cat, mT_an)
    
    ## -------------------------------------------------------------------------------------------------

    mjKcond     = np.zeros ([mjCaMax, mjAnMax+1])
    mnCat_Kcond = np.zeros([3,7,7])
    divKcond    = np.zeros([4,10, 6])
    trivKcond   = np.zeros([4,20, 6])
    
    ## -------------------------------------------------------------------------------------------------
    ## Convert thermodynamic constants to condtional constants 
    
    # for in = 1:maxr
    # Y1 = mjCaMax *(in-1)
    # Y2 = 7  *(in-1)
    # Y3 = 10 *(in-1) 
    # Y4 = 20 *(in-1)
    
    ## Major Ions
    for cat in range(mjCaMax):
        for an in range(mjAnMax+1):
            if mjKthermo[cat,an] !=0:
                mjKcond [cat,an] =  mjKthermo[cat,an] * ((gF_cat[:,cat] * gF_an[:,an])/gmjIP[cat,an])
          
    
    ## Minor Cations
    Lig = np.zeros([1,7])
    Lig = [1,2,3,4,5,6,8]           #[Cl, SO4, CO3, HCO3, Br, F, OH]

    C = np.zeros([1,7])
    C = [6,7,8,9,10,11,22]


    for II in range(7):
        for  JJ in range(7):
            for  KK in range(3): 
                if mnCat_Kthermo[KK,II,JJ] !=0:
                    mnCat_Kcond[KK,II,JJ] = mnCat_Kthermo[KK,II,JJ]*(gF_cat[0][C[II]-1] * (gF_an[0][Lig[JJ]-1]**(KK+1))/mnCatIP[KK,II,JJ])
    
    ## Convert thermodynamic constants to condtional constants 
    ## DIVALENTS 
        L = np.zeros([1,6])
        L = [1,2,3,4,6,8]             # [Cl, SO4, CO3, HCO3, F, OH]
    
        for II in range(10):
            for  JJ in range(6):
                for  KK in range(4): 
                    if divKthermo[KK,II,JJ] !=0:
                        divKcond[KK,II,JJ] = divKthermo[KK,II,JJ] * ((gF_cat[0][11 + II])*(gF_an[0][L[JJ]-1]**(KK+1))/gdivIP[KK,II,JJ])



    ## TRIVALENTS ASSOCIATION CONSTANTS
        for II in range(20):
            for  JJ in range(6):
                for  KK in range(4):
                    if trivKthermo[KK,II,JJ] !=0:
                        trivKcond [KK,II,JJ] = trivKthermo[KK,II,JJ] * ((gF_cat[0][22 + II])*(gF_an[0][Lig[JJ]-1]**(KK+1))/gtrivIP[KK,II,JJ])

    return mjKcond, mnCat_Kcond, divKcond, trivKcond
```
