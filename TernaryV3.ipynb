```python
# ternaryV3   This function calculates mixing parameter, theta_Mc and theta_Xa
#             (which account for interactions of like-charged ions) and
#             triplet interaction parameters, phi_AAC and phi_CCA, at a
#             given temperature. The function ouputs these as arrays.
# 
#             These interaction parameters are needed to calculate free activity 
#             coefficients using Pitzer equations 
#                          *******************************
#             [theta_Xa, theta_Mc, phi_CCA, phi_AAC] = ternaryV3(T)
#                              T = Temperature in Kelvin
#                             NB:- one value of T at a time.
#                          *******************************   
#             Data from Table A10 and A11 in Millero and Pierrot (1998) unless otherwise 
#             specified.
#
#             See also  binaryV3, gammaFV3

## ------------------------------------------------------------------------------
##              NUMBERING SYSTEM
#               Major Cations    
#               Na = 0;       K = 1;     Mg = 2;       Ca = 3;      Sr = 4
#
#               ************************************************************************************
#
#               Minor Cations
#                H = 5;      Li = 6;      Rb = 7;       Cs = 8;      NH4 = 9;    Ba = 10;   Mn = 11;    
#           Fe(II)= 12;     Co = 13;     Ni = 14;   Cu(II)= 15;      Zn = 16;   UO2 = 17;   Be = 18;    
#              Cd = 19;  Pb(II)= 20;   Cu(I)= 21;      La = 22;      Ce = 23;    Pr = 24;   Nd = 25;      
#              Pm = 26;     Sm = 27;     Eu = 28;      Gd = 29;      Tb = 30;    Dy = 31;   Ho = 32;      
#              Er = 33;     Tm = 34;     Yb = 35;      Lu = 36;       Y = 37;    Al = 38;   Ga = 39;      
#              In = 40; Fe(III)= 41;
#
#               ************************************************************************************
#
#               Major Anions   
#               Cl = 0;     SO4 = 1;     CO3 = 2;     HCO3 = 3;    Br = 4;     F = 5   B(OH)4 = 6
# 
#               ************************************************************************************
#
#              Minor Anions
#                  OH = 7;   HSO4 = 8;     HS = 9;      I = 10;    ClO3  = 11;  ClO4 = 12;    BrO3 = 13;     
#                CNS = 14;   NO2 = 15;   NO3 = 16;  H2PO4 = 17;     HPO4 = 18;   PO4 = 19;  H2AsO4 = 20;  
#              HAsO4 = 21;  AsO4 = 22;  HSO3 = 23;    SO3 = 24;  Acetate = 25;
#          
#               ************************************************************************************
#            
#              See also  tenaryV3, gammaFV3
## ------------------------------------------------------------------------------

##  Assign values to global variables
mjCaMax = 5            #In python array goes 0,1,2,3,4,5,6, so mjCaMax must go to 7 to include the 6th variable
mnCaMax = 36
CaMax = mjCaMax + mnCaMax

mjAnMax = 7            #In python array goes 0,1,2,3,4,5,6,7,8 so mjCaMax must go to 8 to include the 7th variable
mnAnMax = 18
AnMax = mjAnMax + mnAnMax

NeutMax = 7

## ------------------------------------------------------------------------------
def ternaryV3(T): 
## ------------------------------------------------------------------------------
## THETA
## Theta Positive ions
    theta_Mc = np.zeros([CaMax, mjCaMax]) 

## Na
# N/A                                                # Na-Na (0,0)
    theta_Mc[0,1] = -5.02312111e-2 + 14.0213141/T    # Na-K  (0,1)
    theta_Mc[0,2] = 0.07                             # Na-Mg (0,2)
    theta_Mc[0,3] = 0.05                             # Na-Ca (0,3)
    theta_Mc[0,4] = 0.07                             # Na-Sr (0,4)

## K
    theta_Mc[1,0] = theta_Mc[0,1]                    # K-Na  (1,0)
# N/A                                                # K-K   (1,1)
# Theta_Mc(1,2) = 0.0;                               # K-Mg  (1,2)
    theta_Mc[1,3] = 0.1156                           # K-Ca  (1,3)
    theta_Mc[1,4] = 0.01                             # K-Sr  (1,4)

## Mg
    theta_Mc[2,0] = theta_Mc[0,2]                    # Mg-Na (2,0)
# Theta_Mc(2,1) = Theta_Mc(1,2);                     # Mg-K  (2,1)
# N/A                                                # Mg-Mg (2,2)
    theta_Mc[2,3] = 0.007                            # Mg-Ca (2,3)
# Theta_Mc(2,4) = 0.0;                               # Mg-Sr (2,4)

## Ca
    theta_Mc[3,0] = theta_Mc[0,3]                    # Ca-Na (3,0)
    theta_Mc[3,1] = theta_Mc[1,3]                    # Ca-K  (3,1)
    theta_Mc[3,2] = theta_Mc[2,3]                    # Ca-Mg (3,2)
# N/A                                                # Ca-Ca (3,3)
# Theta_Mc(3,4) = 0.0;                               # Ca-Sr (3,4)

## Sr
    theta_Mc[4,0]  = theta_Mc[0,4]                   # Sr-Na (4,0)
    theta_Mc[4,1]  = theta_Mc[1,4]                   # Sr-K  (4,1)
# Theta_Mc(4,2) = Theta_Mc(2,4);                     # Sr-Mg (4,2)
# Theta_Mc(4,3) = Theta_Mc(3,4);                     # Sr-Ca (4,3)
# N/A                                                # Sr-Sr (4,4)

## H
    theta_Mc[5,0]  = 0.03416 - 2.09e-4*(T-298.15)    # H-Na  (5,0)
    theta_Mc[5,1]  = 0.005 - 2.275e-4*(T-298.15)     # H-K   (5,1)
    theta_Mc[5,2]  = 0.062 + 3.275e-4*(T-298.15)     # H-Mg  (5,2)
    theta_Mc[5,3]  = 0.0612 + 3.275e-4*(T-298.15)    # H-Ca  (5,3)
    theta_Mc[5,4]  = 0.0591 + 4.5e-4*(T-298.15)      # H-Sr  (5,4)

## Li = 0

## Rb = 0

## Cs
    theta_Mc[8,0]  = -0.033                          # Cs-Na (8,0)

## NH4 = 0

## Ba
    theta_Mc[10,0] = 0.067                           # Ba-Na (10,0)
    theta_Mc[10,1] = 0.01                            # Ba-K  (10,1)

## Mn
    theta_Mc[11,0] = 0.082                           # Ba-Na (11,0)

## Fe(II) T Fe(III) = 0.0
## ------------------------------------------------------------------------------
## Theta Negative ions
    theta_Xa = np.zeros([AnMax+1, mjAnMax])

## Cl  
# N/A                                                # Cl-Cl         (0,0)
    theta_Xa[0,1] = 0.07                             # Cl-SO4        (0,1)
    theta_Xa[0,2] = -0.053                           # Cl-CO3        (0,2)
    theta_Xa[0,3] = 0.0359                           # Cl-HCO3       (0,3)
# Theta_Xa(0,4)  = 0                                 # Cl-Br         (0,4)
    theta_Xa[0,5] = 0.01                             # Cl-F          (0,5)
    theta_Xa[0,6] = -0.0323 - 4.2333e-5*(T-298.15) - 2.1926e-5*(T-298.15)**2 
                                                     # Cl-B(OH)4     (0,6)
## SO4 
    theta_Xa[1,0] = theta_Xa[0,1]                    # SO4-Cl        (1,0)
# N/A                                                # SO4-SO4       (1,1)
    theta_Xa[1,2] = 0.02                             # SO4-CO3       (1,2)
    theta_Xa[1,3] = 0.01                             # SO4-HCO3      (1,3)
# Theta_Xa(1,4)  = 0                                 # SO4-Br        (1,4)
# Theta_Xa(1,5)  = 0                                 # SO4-F         (1,5)
    theta_Xa[1,6] = -0.012                           # SO4-B(OH)4    (1,6)

## CO3    
    theta_Xa[2,0] = theta_Xa[0,2]                    # CO3-Cl        (2,0)
    theta_Xa[2,1] = theta_Xa[1,2]                    # CO3-SO4       (2,1)

## HCO3     
    theta_Xa[3,0] = theta_Xa[0,3]                    # HCO3-Cl       (3,0)
    theta_Xa[3,1] = theta_Xa[1,3]                    # HCO3-SO4      (3,1)

## Br  = 0

## F   
    theta_Xa[5,0] = theta_Xa[0,5]                    # F-Cl          (5,0)

## B(OH)4 
    theta_Xa[6,0] = theta_Xa[0,6]                    # B(OH)4-Cl     (6,0)
    theta_Xa[6,1] = theta_Xa[1,6]                    # B(OH)4-SO4    (6,1)

## OH 
    theta_Xa[7,0] = -0.05 + 3.125e-04*(T-298.15) - 8.362e-06*(T-298.15)**2    # OH-Cl  (7,0)             
    theta_Xa[7,1] = -0.013                           # OH-SO4        (7,1)
    theta_Xa[7,2] = 0.1                              # OH-CO3        (7,2)  ... http://www.aim.env.uea.ac.uk/aim/accent4/parameters.html  
# Theta_Xa(7,3) = 0;                                 # OH-HCO3       (7,3)
    theta_Xa[7,4] = -0.065                           # OH-Br         (7,4)
# Theta_Xa(7,5) = 0                                  # OH-F          (7,5)
# Theta_Xa(7,6) = 0                                  # OH-B(OH)4     (7,6)

## HSO4
    theta_Xa[8,0] = -0.006                           # HSO4-Cl       (8,0)

## HS  = 0 
## I    = 0  
## ClO3 = 0       
## ClO4 = 0  
## BrO3 = 0     
## CNS  = 0
## NO2  = 0  
## NO3   
    theta_Xa[16,0] = 0.016                           # NO3-Cl       (16,0)

## H2PO4   
    theta_Xa[17,0] = 0.1                             # H2PO4-Cl     (17,0)

## HPO4  
    theta_Xa[18,0] = -0.105                          # HPO4-Cl      (18,0)

## PO4   
    theta_Xa[19,0] = -0.59                           # PO4-Cl       (19,0)

## H2AsO4
    theta_Xa[20,0] = 0.228                           # H2AsO4-Cl    (20,0)

## HAsO4 
    theta_Xa[21,0] = 0.122                           # HAsO4 -Cl    (21,0)

## AsO4  
    theta_Xa[22,0] = 0.06                            # AsO4-Cl      (22,0)

## HSO3 = 0

## SO3   
    theta_Xa[24,0] = 0.0991                          # SO3-Cl       (24,0)

## Acetate 
    theta_Xa[25,0] = -0.017                          # Acetate-Cl   (25,0)
## ********************************************************************************************************************
## Phi_CCA holds the values for cation-cation-anion 
# matlab = [row, integer along, set]
# python = [set, row, integer along] (keep in mind python starts at 0)

    phi_CCA = np.zeros([mjAnMax, CaMax, mjCaMax+1])

## Na-K-Cl
    phi_CCA[0,0,1] = 1.34211308E-2 - 5.10212917/T
    phi_CCA[0,1,0] = phi_CCA[0,0,1]

## Na-K-SO4
    phi_CCA[1,0,1] = 3.48115174E-2 - 8.21656777/T
    phi_CCA[1,1,0] = phi_CCA[1,0,1]

## Na-K-Br
    phi_CCA[4,0,1] = -0.0022
    phi_CCA[4,1,0] = phi_CCA[4,0,1]

## Na-Mg-Cl
    phi_CCA[0,0,2] = 0.0199 - 9.51/T
    phi_CCA[0,2,0] = phi_CCA[0,0,2]

## Na-Ca-Cl
    phi_CCA[0,0,3] = -7.6398 -1.2990e-2*T + 1.1060e-5*T**2 + 1.8475*log(T)  # Spencer et al 1990 # -0.003
    phi_CCA[0,3,0] = phi_CCA[0,0,3]

## Na-Ca-SO4
    phi_CCA[1,0,3] = -0.012
    phi_CCA[1,3,0] = phi_CCA[1,0,3]

## K-Mg-Cl
    phi_CCA[0,1,2] = 0.02586 - 14.27/T
    phi_CCA[0,2,1] = phi_CCA[0,1,2]

## K-Ca-Cl
    phi_CCA[0,1,3] = 0.047627877 - 27.0770507/T
    phi_CCA[0,3,1] = phi_CCA[0,1,3]

## K-Ca-SO4
# phi_CCA(1,3,1) = 0.0
# phi_CCA(3,1,1) = phi_CCA(1,3,1)

## H-Sr-Cl
    phi_CCA[0,5,4] = 0.0054 - 2.1*1E-4*(T-298.15)
    phi_CCA[0,4,5] = phi_CCA[0,5,4]

## H-Mg-Cl
    phi_CCA[0,5,2] = 0.001 - 7.325*1E-4*(T-298.15)
    phi_CCA[0,2,5] = phi_CCA[0,5,2]

## H-Ca-Cl
    phi_CCA[0,5,3] = 0.0008 - 7.25*1E-4*(T-298.15)
    phi_CCA[0,3,5] = phi_CCA[0,5,3]

## Sr-Na-Cl
    phi_CCA[0,4,0] = -0.015
    phi_CCA[0,0,4] = phi_CCA[0,4,0]

## Sr-K-Cl
    phi_CCA[0,4,1] = -0.015
    phi_CCA[0,1,4] = phi_CCA[0,4,1]

## Na-Mg-SO4
    phi_CCA[1,0,2] = -0.015
    phi_CCA[1,2,0] = phi_CCA[1,0,2]

## K-Mg-SO4
    phi_CCA[1,1,2] = -0.048
    phi_CCA[1,2,1] = phi_CCA[1,1,2]

## Mg-Ca-Cl
    phi_CCA[0,2,3] = 4.15790220e1 + 1.30377312e-2*T -9.81658526e2/T -7.4061986*log(T)  # Spencer et al 1990 -0.012
    phi_CCA[0,3,2] = phi_CCA[0,2,3]

## Mg-Ca-SO4
    phi_CCA[1,2,3] = 0.024
    phi_CCA[1,3,2] = phi_CCA[1,2,3]

## H-Na-Cl
    phi_CCA[0,5,0] = 0.0002

## H-Na-SO4
# Phi_CCA(5,0,1) = 0.0

## H-K-Cl
    phi_CCA[0,5,1] = -0.011

## H-K-SO4
    phi_CCA[1,5,1] = 0.197

## H-Na-Br
    phi_CCA[4,5,0] = -0.012

## H-K-Br
    phi_CCA[4,5,1] = -0.021

## H-Mg-Br
    phi_CCA[4,5,2] = -0.005

## Cs-Na-Cl
    phi_CCA[0,8,0] = -0.003

## Cs-K-Cl
    phi_CCA[0,8,1] = -0.0013

## Ba-Na-Cl
    phi_CCA[0,10,0] = -0.012

## Ba-K-Cl
    phi_CCA[0,10,1] = -0.017

## Mn-Na-Cl
    phi_CCA[0,11,0] = -0.0174
## ********************************************************************************************************************
## Phi_AAC holds the values for anion-anion-cation
# matlab = [row, integer along, set]
# python = [set, row, integer along] (keep in mind python starts at 0)

    phi_AAC = np.zeros([mjCaMax, AnMax, mjAnMax]) 

## Cl-SO4-Na
    phi_AAC[0,0,1] = -0.009
    phi_AAC[0,1,0] = phi_AAC[0,0,1]

## Cl-SO4-K
    phi_AAC[1,0,1] = -0.21248147 + 37.5619614/T + 2.8469833*1E-3*T
    phi_AAC[1,1,0] = phi_AAC[1,0,1]

## Cl-SO4-Ca
    phi_AAC[3,0,1] = -0.018
    phi_AAC[3,1,0] = phi_AAC[3,0,1]

## Cl-CO3-Na
    phi_AAC[0,0,2] = 0.016
    phi_AAC[0,2,0] = phi_AAC[0,0,2]

## Cl-HCO3-Na
    phi_AAC[0,0,3] = -0.0143
    phi_AAC[0,3,0] = phi_AAC[0,0,3]

## Cl-BOH4-Na
    phi_AAC[0,0,6] = -0.0132
    phi_AAC[0,6,0] = phi_AAC[0,0,6]

## Cl-BOH4-Mg
    phi_AAC[2,0,6] = -0.235
    phi_AAC[2,6,0] = phi_AAC[2,0,6]

## Cl-BOH4-Ca
    phi_AAC[3,0,6] = -0.8
    phi_AAC[3,6,0] = phi_AAC[3,0,6]

## Cl-F-Na
    phi_AAC[0,0,5] = 0.0023
    phi_AAC[0,5,0] = phi_AAC[0,0,5]

## HSO4-SO4-Na
# Phi_AAC(8,1,0) = 0.0;
# Phi_AAC(1,8,0) = Phi_AAC(8,1,0);

## CO3-HCO3-Na
# Phi_AAC(2,3,0) = 0.0;
# Phi_AAC(3,2,0) = Phi_AAC(2,3,0);

## CO3-HCO3-K
# Phi_AAC(1,3,2) = 0.0;
# Phi_AAC(3,2,1) = Phi_AAC(2,3,1);

## Cl-SO4-Mg
    phi_AAC[2,0,1] = -0.004
    phi_AAC[2,1,0] = phi_AAC[2,0,1]

## Cl-HCO3-Mg
    phi_AAC[2,0,3] = -0.0196
    phi_AAC[2,3,0] = phi_AAC[2,0,3]

## SO4-CO3-Na
    phi_AAC[0,2,1] = -0.005
    phi_AAC[0,1,2] = phi_AAC[0,2,1]

## SO4-CO3-K
    phi_AAC[1,1,2] = -0.009
    phi_AAC[1,2,1] = phi_AAC[1,1,2]

## SO4-HCO3-Na
    phi_AAC[0,1,3] = -0.005
    phi_AAC[0,3,1] = phi_AAC[0,1,3]

## SO4-HCO3-Mg
    phi_AAC[2,1,3] = -0.161
    phi_AAC[2,3,1] = phi_AAC[2,1,3]

## OH-Cl-Na
    phi_AAC[0,7,0] = -0.006

## OH-Cl-K
    phi_AAC[1,7,0] = -0.006

## OH-Cl-Ca
    phi_AAC[3,7,0] = -0.025

## OH-SO4-Na
    phi_AAC[0,7,1] = -0.009

## OH-SO4-K
    phi_AAC[1,7,1] = -0.05

## OH-Br-Na
    phi_AAC[0,7,4] = -0.018

## OH-Br-K
    phi_AAC[1,7,4] = -0.014

## HSO4-Cl-Na
    phi_AAC[0,8,0] = -0.006

## HSO4-SO4-Na
# phi_AAC(9,2,1) = 0

## HSO4-SO4-K
    phi_AAC[1,8,1] = -0.0677

## NO3-Cl-Na
    phi_AAC[0,16, 0] = -0.006

## NO3-Cl-K

## H2PO4-Cl-Na
    phi_AAC[0,17, 0] = -0.028            # Hershey et al.1989 '0 'Millero 1982 thalassia
    
## H2PO4-Cl-K
# phi_AAC(18, 1, 1) = 0                   # co2brine   '-.0105  'Pitzer Silvester 1976

## HPO4-Cl-Na 
    phi_AAC[0,18, 0] = -0.003            # Hershey et al.1989'0   'Millero 1982 thalassia

## PO4-Cl-Na  
    phi_AAC[0,19, 0] = 0.11              # Hershey et al.1989 '0  'Millero 1982 thalassia

## H2As04-Cl-Na  
# phi_AAC(21, 1, 1) = 0                   # Millero 1982 thalassia

## HAs04-Cl-Na  
# phi_AAC(22, 1, 1) = 0                   # Millero 1982 thalassia

## As04-Cl-Na  
# phi_AAC(23, 1, 1) = 0                   # Millero 1982 thalassia

## SO3-Cl-Na
    phi_AAC[0,24, 0] = -0.0156
    
    return theta_Mc, theta_Xa, phi_CCA, phi_AAC
```
